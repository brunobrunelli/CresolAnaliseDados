```python
#Imports:
import gzip
import pandas as pd
from datetime import datetime
```

```python
# Lógica de leitura dos dados do arquivo de Logs:
with gzip.open("../FonteDados/apache.log.gz","rb") as f:
    data = f.read()
```

```python
# Decodificando os dados de entrada para "utf-8":
dataLog = data.decode("utf-8")
```

```python
# Criamos uma lista para remover os Caracteres indesejáveis em nossos dados de entrada:
lista_remove = ['"',')','(',';','[',']']
```

```python
# Removendo os caracteres dos dados de entrada: 
for i in lista_remove:
    dataLog = dataLog.replace('%s' %i,'')
```

```python
# Recortamos os dados de entrada, dividindo cada pedaço da string em substrings dentro de uma lista:  
result = [x.strip() for x in dataLog.split('\r\n')]
```

```python
# Removemos valores em branco encontrados na lista de valores:
for i in result:
    if i == '':
        result.remove('') 
```

```python
listaLog = []
```

```python
for i in range(len(result)):
    lista = []
    user_agent_lista = []
    
    try:
        aux = result[i].split(' ')
        
        # Lógica para percorrer as 9 primeiras colunas, inserindo na lista:
        for f in range(0,10,1):
            # Lógica para remover a coluna '-':
            if f != 1: 
                # Lógica para incluir duas novas colunas (DATE e TIME):
                if f == 4: 
                    #'%Y-%M-%DT%H:%M:%SZ'
                    lista.append(datetime.strptime(aux[f],'%Y-%m-%dT%H:%M:%SZ').strftime('%Y-%m-%d'))
                    lista.append(datetime.strptime(aux[f],'%Y-%m-%dT%H:%M:%SZ').strftime('%H:%M:%S'))
                else:
                    lista.append(aux[f])

        # Lógica para concatenar as últimas colunas, dando origem a coluna USER_AGENT:
        for n in range(10,len(aux),1):
                user_agent_lista.append(aux[n])

        user_agent = ' '.join(user_agent_lista)
        lista.append(user_agent)
        
        # Incluímos os valores tratados da lista em uma sublista:
        listaLog.append(lista)
        
        # Excluímos as listas carregadas da memória para desalocar espaço:
        del lista
        del user_agent_lista
        
    except:
        print('Erro: A lista contém valores que não são esperados.')        
```

```python
# Definindo uma lista com as Colunas do DataFrame:
labels = ['IP_ADDRESS','USER','ID_USER','DATE','TIME','METHOD','REQUESTED_RESOURCE','STATUS_CODE','SIZE_OBJECT','REFER','USER_AGENT']
```

```python
# Criação do DataFrame a partir da lista de Logs:
dfLog = pd.DataFrame(listaLog,columns=labels)
```
